- load the libraries

```python
from bs4 import BeautifulSoup # this module helps in web scrapping.'

import requests #to download a webpage
import pandas as pd
```

- implement a request

```python
url = "https://en.wikipedia.org/wiki/List_of_most_popular_given_names"  #web address
```

```python
data  = requests.get(url).text #recover the html page

data[0:1000]  #show some of the content
```

- Creates a "soup" to navigate through the html structure

```python
soup = BeautifulSoup(data,"html.parser")  # create a soup object using the variable 'data'
allTables = soup.findAll('table') #find all the table objects

len(allTables) #return the number of tables
```

```python
#for table in allTables: #shows some of the tables
  #  print(table.prettify())

allTables[0].prettify() #print the content of the table
```

- analise the tables row content by navigating through the `<td>` `<tr>` elements

```python
for i,table in enumerate(allTables): #iterate in all tables
    print(i)
    for rows in table.tbody.find_all('tr'):
        cols = rows.find_all('td')
        for col in cols:
            if col != []:
                print(col.text)
    print("\n")        
        
# rows = table.children #this is the <tr> element    table -> tr (row) -> td (collumn)
```
